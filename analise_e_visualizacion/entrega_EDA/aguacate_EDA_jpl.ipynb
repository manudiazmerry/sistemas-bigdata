```python


def dist_custom(dataset, columns_list, rows, cols, subtitle):
    fig, axs = plt.subplots(rows, cols,figsize=(16,16))
    fig.suptitle(subtitle,y=0.92, size=16)
    axs = axs.flatten()
    for i, data in enumerate(columns_list):
        sns.histplot(dataset[data], ax=axs[i])
        axs[i].set_title(data + ', skewness is '+str(round(dataset[data].skew(axis = 0, skipna = True),2)))
        
dist_custom(dataset=df, columns_list= df.columns[1:10], rows=3, cols=3, subtitle='Distibución para cada variable')
```

```python
#Volumenes de venta según tipo por año (1). Relación de volumnes.

con=df_totalus[df_totalus['Tipo']=='conventional'].groupby('Año')['Volumen_total_vendido'].agg('mean')
org=df_totalus[df_totalus['Tipo']=='organic'].groupby('Año')['Volumen_total_vendido'].agg('mean')
con/org
```

```python
plt.figure(figsize = (8, 6))
plt.plot(con, label='conventional', marker='o')
plt.plot(org, label='organic', marker ='x')
plt.ylabel('Volumen total vendido')
plt.xlabel('Año')
plt.title('Volumen total vendido por tipo')
plt.legend()
plt.show()
```

```python
#Volumenes de venta según tipo por año (2)

numero_grupos = len(con)
indice_barras=np.arange(numero_grupos)
ancho_barras=0.4
plt.bar(indice_barras, con, width=ancho_barras, label='conventional')
plt.bar(indice_barras + ancho_barras, org, width=ancho_barras, label='organic')
plt.xticks(indice_barras + ancho_barras, ('2015', '2016','2017', '2018', '2019', '2020','2021'))

plt.ylabel('Volumen total vendido')
plt.xlabel('Año')
plt.title('Volumen total vendido por tipo')
plt.legend()
plt.show()
```

```python
#evolución precios por años según tipo (1)

con_precio_anios=df_totalus[df_totalus['Tipo']=='conventional'].groupby('Año')['Precio_promedio_aguacate'].agg('mean')
org_precio_anios=df_totalus[df_totalus['Tipo']=='organic'].groupby('Año')['Precio_promedio_aguacate'].agg('mean')
```

```python
#evolución precios por años según tipo (2)

numero_grupos = len(con)
indice_barras=np.arange(numero_grupos)
ancho_barras=0.4
plt.bar(indice_barras, con_precio_anios, width=ancho_barras, label='conventional')
plt.bar(indice_barras + ancho_barras, org_precio_anios, width=ancho_barras, label='organic')
plt.xticks(indice_barras + ancho_barras, ('2015', '2016','2017', '2018', '2019', '2020','2021'))

plt.ylabel('Precio promedio aguacate')
plt.xlabel('Año')
plt.title('Precio promedio por tipo')
plt.legend()
plt.show()
```

```python
sns.lineplot(x='Año',y= con_precio_anios, data=con_precio_anios).set(title= 'Precios aguacate convencional')
plt.show()
```

```python
sns.lineplot(x='Año',y= org_precio_anios, data=org_precio_anios).set(title= 'Precios aguacate orgánico')
plt.show()
```
